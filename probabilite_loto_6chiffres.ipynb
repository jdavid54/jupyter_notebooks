```python

def norep(N,n):
    # espace des valeurs N, nombre de tirages n
    # p probabilité pour aucune répétition
    p = 1
    for i in range(N-n+1,N):
        p *= i/N    
    return p

def prob1r(N,n):
    # une répétition pour n tirages
    return 1-exp(-n*(n-1)/(2*N))
    #return 1-e**(-n*(n-1)/(2*N))

def n_tirage(p,l):    
    #print(y[y>p])
    #print(len(y[y>p]))
    t = l-len(y[y>p])-1
    #t rang en dessous de p
    #print(t)
    #print(y[t:])
    print(t+1, y[t:t+2])
    return t+1
```

```python

#r = 1
N = 1000000
n = 1200
l = 4000

print(prob1r(N,n))
print(1-norep(N,n))

p = .9
x = np.arange(1, l)
y = prob1r(N,x)
t = n_tirage(p,l)
```

```python
# nombres de tirages pour avoir un taux demandé de répétition
plt.plot(x,y)
plt.xlabel(f"Tirages")
plt.plot(t,y[t],'ro')
plt.ylabel(f"Prob. %")
plt.title(f'{t} tirages pour {p*100}% d\'avoir une répétition')
plt.grid()
plt.show()
```
